```python
df = pd.read_csv('tested.csv')
df.head()
```

```python
df.tail()
```

```python
df.sample(5)
```

```python
print("Rows and Columns:", df.shape)
df.info()
df.describe(include='all')
```

```python
df.isnull().sum()
```

```python
(df.isnull().sum()/len(df))*100
```

```python
df['Age'].fillna(df['Age'].median(), inplace=True)
df['Fare'].fillna(df['Fare'].median(), inplace=True)
df['Cabin'].fillna('Unknown', inplace=True)
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
df.duplicated().sum()
```

```python
plt.figure(figsize=(8,4))

sns.boxplot(x=df['Fare'])

plt.title("Outliers in Fare")

plt.show()
```

```python
plt.figure(figsize=(8,4))

sns.boxplot(x=df['Age'])

plt.title("Outliers in Age")

plt.show()
```

```python
df.to_csv('cleaned_titanic.csv', index=False)
```

```python
plt.figure(figsize=(6,4))

df['Survived'].value_counts().plot(kind='bar')

plt.title('Survival Distribution')
plt.xlabel('Survived')
plt.ylabel('Count')

plt.show()
```

```python
plt.figure(figsize=(6,6))

df['Sex'].value_counts().plot(
    kind='pie',
    autopct='%1.1f%%'
)

plt.title('Gender Distribution')

plt.ylabel('')

plt.show()
```

```python
plt.figure(figsize=(6,4))

df['Pclass'].value_counts().plot(kind='bar')

plt.title('Passenger Class Distribution')
plt.xlabel('Class')
plt.ylabel('Count')

plt.show()
```

```python
plt.figure(figsize=(8,5))

plt.hist(df['Age'], bins=20)

plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')

plt.show()
```

```python
plt.figure(figsize=(8,5))

plt.hist(df['Fare'], bins=20)

plt.title('Fare Distribution')
plt.xlabel('Fare')
plt.ylabel('Frequency')

plt.show()
```

```python
sns.countplot(
    x='Sex',
    hue='Survived',
    data=df
)

plt.title('Survival by Gender')

plt.show()
```

Key Insights

1. The dataset contains 418 passenger records.

2. Missing values were found in Age, Fare, and Cabin columns.

3. Missing Age and Fare values were replaced using the median.

4. Missing Cabin values were replaced with "Unknown".

5. Fare column contains several outliers.

6. Most passengers belonged to Passenger Class 3.

7. Female passengers had a higher survival rate than male passengers.

8. Data cleaning improved dataset quality and reliability.
